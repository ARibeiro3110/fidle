## Step 2 - Prepare data
### 2.1 - Get basic dataset

```python
x_train, y_train, x_test, y_test = Loader_MNIST.get(concatenate=False)
```

### 2.2 - Add noise to dataset
Nous allons ajouter du buit a notre dataset en entrée (x_train, x_test)  
Notre objectif est de prédire les données sans bruit (y_train, y_test)

```python
def noise_it(x_data, about):
    for i,image in enumerate(x_data):
        pwk.update_progress(about,i+1,len(x_data))
        image=random_noise(image, mode='gaussian', mean=0, var=0.4)
        image=random_noise(image, mode='s&p',      amount=0.2, salt_vs_pepper=0.5)
        image=random_noise(image, mode='poisson')  # Poisson
        image=random_noise(image, mode='speckle',  mean=0, var=0.1)
        x_data[i]=image
    return x_data


# ---- What we want
#
y_train = np.copy(x_train)
y_test  = np.copy(x_test)

# ---- Add noise to x_train, x_test
#
x_train = noise_it(x_train, 'Add noise to x_train :')
x_test  = noise_it(x_test,  'Add noise to x_test  :')
```

### 2.3 - Have a look

```python
print('Noisy train dataset (x_train) : ',x_train.shape)
print('Clean train dataset (y_train) : ',y_train.shape)
print('Noisy test dataset  (x_test)  : ',x_test.shape)
print('Clean test dataset  (y_test)  : ',y_test.shape)

pwk.plot_images(y_train[:5], None, indices='all', columns=5, x_size=3,y_size=3, save_as='01-original')
pwk.plot_images(x_train[:5], None, indices='all', columns=5, x_size=3,y_size=3, save_as='02-noisy')
```

## Step 3 - Build model

```python
latent_dim = 10

# X = tf.ones([10,10,10]) * 100
# Y = noiseLayer(mean = 0, std = 0.1)(X)
```

## Step 4 - Instanciate and compile model

#### Encoder

```python
reload(modules.AE)
from modules.AE          import AE

inputs    = keras.Input(shape=(28, 28, 1))
x         = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(inputs)
x         = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
x         = layers.Flatten()(x)
x         = layers.Dense(16, activation="relu")(x)
z         = layers.Dense(latent_dim)(x)

encoder = keras.Model(inputs, z, name="encoder")
# encoder.summary()
```

#### Decoder

```python
inputs  = keras.Input(shape=(latent_dim,))
x       = layers.Dense(7 * 7 * 64, activation="relu")(inputs)
x       = layers.Reshape((7, 7, 64))(x)
x       = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
x       = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
outputs = layers.Conv2DTranspose(1, 3, activation="sigmoid", padding="same")(x)

decoder = keras.Model(inputs, outputs, name="decoder")
# decoder.summary()
```

#### AE


```python
ae = AE(encoder, decoder)

ae.compile(optimizer=keras.optimizers.Adam(), loss='binary_crossentropy')
```

## Step 5 - Train
20' on a CPU  
1'12 on a GPU (V100, IDRIS)

```python
# ---- Callback : Images
pwk.mkdir(f'{run_dir}/images')
filename = run_dir + "/images/image-{epoch:03d}-{i:02d}.jpg"
callbacks_images = ImagesCallback2(filename, x_test=x_test, z_dim=latent_dim, encoder=encoder,decoder=decoder)

# ---- Callback : Best model
pwk.mkdir(f'{run_dir}/models')
filename = run_dir + "/models/best_model"
callback_bestmodel = BestModelCallback(filename)

# ---- Callback tensorboard
dirname = run_dir + "/logs"
callback_tensorboard = TensorBoard(log_dir=dirname, histogram_freq=1)

callbacks_list = [callbacks_images, callback_bestmodel, callback_tensorboard]
```

```python
%%time

n          = 6000
batch_size = 128
epochs     = 15

history = ae.fit(x_train[:n],y_train[:n], epochs=epochs, batch_size=batch_size, callbacks=callbacks_list)
```

## Step 6 - History

```python
pwk.plot_history(history,  plot={"Loss":['loss']}, save_as='03-history')
```

## Step 7 - Learning progression

```python
imgs=[]
labels=[]
for epoch in range(15):
    for i in range(5):
        filename = run_dir + "/images/image-{epoch:03d}-{i:02d}.jpg".format(epoch=epoch, i=i)
        img      = io.imread(filename)
        imgs.append(img)
        

pwk.subtitle('Real images :')
pwk.plot_images(y_test[:5], None, indices='all', columns=5, x_size=2,y_size=2, save_as='04-original-real')

pwk.subtitle('Noisy images :')
pwk.plot_images(x_test[:5], None, indices='all', columns=5, x_size=2,y_size=2, save_as='05-original-noisy')

pwk.subtitle('Learning...')
pwk.plot_images(imgs, None, indices='all', columns=5, x_size=2,y_size=2, save_as='06-learning')

pwk.subtitle('Noisy images :')
pwk.plot_images(x_test[:5], None, indices='all', columns=5, x_size=2,y_size=2, save_as=None)

pwk.subtitle('Real images :')
pwk.plot_images(y_test[:5], None, indices='all', columns=5, x_size=2,y_size=2, save_as=None)
```
