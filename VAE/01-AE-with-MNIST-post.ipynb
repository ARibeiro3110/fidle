## Step 1 - Init python stuff

```python
import numpy as np
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
import scipy.stats
import sys, importlib
import h5py
import random

import matplotlib
import matplotlib.pyplot as plt

from modules.AE import AE

sys.path.append('..')
import fidle.pwk as pwk


run_dir = './run/AE-MNIST.001'
datasets_dir = pwk.init('VAE2', run_dir)
```

## Step 2 - Get data

```python
with  h5py.File('./data/mnist-noisy.h5','r') as f:
    x_test  = f['x_test'][:]
    y_test  = f['y_test'][:]
    print('Loaded.')
    print('x_test shape is : ',x_test.shape)
    print('y_test shape is : ',y_test.shape)
```

## Step 3 - Reload best model

```python
ae=AE()
ae.reload(f'{run_dir}/models/best_model')
```

## Step 4 - Image reconstruction

```python
y_pred = ae.predict(x_test)
y_pred.shape

# cf https://keras.io/api/models/model/
```

## Step 5 - Latent space

```python
i=random.randint(0,len(y_pred)-8)
j=i+8

pwk.subtitle('Noisy images :')
pwk.plot_images(x_test[i:j], None, indices='all', columns=8, x_size=2,y_size=2, save_as='03-original-noisy')

pwk.subtitle('Predict images :')
pwk.plot_images(y_pred[i:j], None, indices='all', columns=8, x_size=2,y_size=2, save_as='04-predict')

pwk.subtitle('Real images :')
pwk.plot_images(y_test[i:j], None, indices='all', columns=8, x_size=2,y_size=2, save_as='05-original-real')

```
